```python
from pathlib import Path


def find_project_root(start: Path) -> Path:
    for path in [start, *start.parents]:
        if (path / "dataset" / "train.json").exists():
            return path
    raise FileNotFoundError("Could not find project root containing dataset/train.json")


PROJECT_ROOT = find_project_root(Path.cwd().resolve())
DATASET_DIR = PROJECT_ROOT / "dataset"

TRAIN_PATH = DATASET_DIR / "train.json"
TEST_PATH = DATASET_DIR / "test.json"

assert TRAIN_PATH.exists(), f"Missing training data: {TRAIN_PATH}"
assert TEST_PATH.exists(), f"Missing test data: {TEST_PATH}"

print(f"Project root: {PROJECT_ROOT}")
print(f"Train path: {TRAIN_PATH}")
```

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("ml-challenge-eda")
    .master("local[*]")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("WARN")

print(f"Spark version: {spark.version}")
print(f"Spark UI: {spark.sparkContext.uiWebUrl}")
```

```python
df = spark.read.option("multiLine", True).json(str(TRAIN_PATH))
df.show()
```
